```python
from sklearn.cluster import KMeans  
 
wcss_list= []  #Initializing the list for the values of WCSS  
 
#Using for loop for iterations from 1 to 10.  
 
for i in range(1, 11):  
 
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state= 42)  
 
    kmeans.fit(X_Train_scaled)  
 
    wcss_list.append(kmeans.inertia_)  
 
plt.plot(range(1, 11), wcss_list)  
 
plt.title('The Elobw Method Graph')  
 
plt.xlabel('Number of clusters(k)')  
 
plt.ylabel('wcss_list')  
 
plt.show()
```

```python
#training the K-means model on a dataset  
kmeans = KMeans(n_clusters=3, init='k-means++', random_state= 42)  
y_predict= kmeans.fit_predict(X_Train_scaled) 
```

```python
#visulaizing the clusters  
plt.scatter(X_Train_scaled[y_predict == 0, 0], X_Train_scaled[y_predict == 0, 1], s = 10, c = 'blue', label = 'Cluster 1') #for first cluster  
plt.scatter(X_Train_scaled[y_predict == 1, 0], X_Train_scaled[y_predict == 1, 1], s = 10, c = 'black', label = 'Cluster 2') #for second cluster  
plt.scatter(X_Train_scaled[y_predict== 2, 0], X_Train_scaled[y_predict == 2, 1], s = 10, c = 'red', label = 'Cluster 3') #for third cluster   
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 30, c = 'yellow', label = 'Centroid')   
plt.title('AGE')    
plt.ylabel('DEPENDENTS')
#plt.title('Clusters of customers')    
#plt.ylabel('Spending Score (1-10)')  
plt.legend()  
plt.show()  
```
